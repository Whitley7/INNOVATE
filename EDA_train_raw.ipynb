# EDA for RAW train dataset

### 1. IMPORTS + LOAD RAW TRAIN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

train_raw = pd.read_csv("train.csv", parse_dates=["datetime"])
train_raw.head()
```

### 2| INFO + SHAPE + HOUSEHOLDS

```python
train_raw.info()
print("\nShape:", train_raw.shape)
print("\nHousehold counts:")
train_raw["home_id"].value_counts()
```

### 3| TIME RANGE PER HOUSEHOLD

```python
train_raw.groupby("home_id")["datetime"].agg(["min", "max"])
```

### 4| SAMPLING FREQUENCY

```python
train_raw["datetime"] = pd.to_datetime(train_raw["datetime"], utc=True, errors="coerce")
train_raw.dtypes
```

```python
raw_sorted = train_raw.sort_values(["home_id", "datetime"])
```

```python
deltas_raw = raw_sorted.groupby("home_id")["datetime"].diff().dt.total_seconds()

deltas_raw.dropna().value_counts().head(20)
```
